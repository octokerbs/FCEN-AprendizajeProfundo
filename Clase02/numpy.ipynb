```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
M = np.array([[3, -9, 0, 5],
              [2, -5, -3, 1],
              [-1, 5, 8, 4]])

A = np.array([-1, 1, 2])
B = np.array([-4, 2, 1, -1])
```

```python
C = np.dot(A, M)
print(C)
```

```python
A = A.reshape(len(A), 1)    # Por default las dimensiones de los numpy array son (n,) en vez de (n,1)
B = B.reshape(len(B), 1)
M2 = A @ B.T
print(M2)
```

```python
D = np.dot(M, B)
print(D)
```

```python
a = np.dot(A.T, A)
print(a)
```

```python
def f1(x):
    return 1/1 + math.exp(-x)

interval = np.arange(-2, 2, 0.1)
plt.plot(interval, list(map(f1, interval)))
plt.show()
```

```python
def f2(x):
    return math.tanh(x)

interval = np.arange(-2, 2, 0.1)
plt.plot(interval,list(map(f2, interval)))
plt.show()
```

```python
def f3(x):
    return math.sinh(x)

interval = np.arange(-2, 2, 0.1)
plt.plot(interval,list(map(f3, interval)))
plt.show()
```

```python
def f4(x):
    return f1(x) * (1 - f1(x))

interval = np.arange(-2, 2, 0.1)
plt.plot(interval,list(map(f4, interval)))
plt.show()
```

```python
def f5(x):
    return 1 - f2(x)**2

interval = np.arange(-2, 2, 0.1)
plt.plot(interval,list(map(f5, interval)))
plt.show()
```
