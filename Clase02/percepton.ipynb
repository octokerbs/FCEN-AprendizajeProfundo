### Perceptron discreto

```python
class PerceptronSimple(object):
    def __init__(self, n):
        self.w_ = np.zeros(n+1)

    def train(self, X, d, eta, epochs):
        errors = []
        for _ in range(epochs):
            accumulated_error = 0
            for X_i, target in zip(X, d):
                update = eta * (target - self.predict(X_i))
                self.w_[1:] +=  update * X_i
                self.w_[0] +=  update * 1
                accumulated_error += int(update != 0.0)
            errors.append(accumulated_error)
        return errors

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```
